```python
from thinkdsp import *
class Mytri(Sinusoid):
    def __init__(self,*parg, **kargs):
        super(Mytri,self).__init__(*parg, **kargs)
    def evaluate(self,ts):
        cycles = self.freq * ts +self.offset/PI2
        frac,_ = np.modf(cycles)
        ys = frac
        #ys = np.abs(frac - 0.5)
        ys = normalize(unbias(ys), self.amp)
        return ys

sig = Mytri(200)
sig.plot()
```
