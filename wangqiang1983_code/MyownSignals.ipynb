```python
from thinkdsp import *
class Mytri(Sinusoid):
    def __init__(self,*parg, **kargs):
        super(Mytri,self).__init__(*parg, **kargs)
    def evaluate(self,ts):
        cycles = self.freq * ts +self.offset/PI2
        frac,_ = np.modf(cycles)
        ys = frac
        #ys = np.abs(frac - 0.5)
        ys = normalize(unbias(ys), self.amp)
        return ys

sig = Mytri(200)
sig.plot()
```

```python

class Mysin(Sinusoid):
    def __init__(self,*parg, **kargs):
        super(Mysin,self).__init__(*parg, **kargs)
        
    def evaluate(self,ts):
        cycles = self.freq * ts +self.offset/PI2
        frac,_ = np.modf(cycles)
        ys = np.sin(frac*PI2)
        #ys = np.abs(frac - 0.5)
        #ys = normalize(unbias(ys), self.amp)
        return ys
    
mysin = Mysin(5)
mysin.plot()
```

```python
mix = 0
minif = 5
for i in range(5):
    f = minif*(i+1)
    sig =  Mysin(f)
    sig.plot()
    mix += sig
mix.plot()
```

```python
ww = mix.make_wave()
ww.plot()
```

```python
ss = ww.make_spectrum()
ss.plot(high=100)
```

```python

class Mysquare(Sinusoid):
    def __init__(self,*parg, **kargs):
        super(Mysquare,self).__init__(*parg, **kargs)
        
    def evaluate(self,ts):
        cycles = self.freq * ts +self.offset/PI2
        frac,_ = np.modf(cycles)
        ys = np.sign(unbias(frac))
        #ys = np.abs(frac - 0.5)
        #ys = normalize(unbias(ys), self.amp)
        return ys
    
mysquare = Mysquare(5)
mysquare.plot()
```

```python
ww = mysquare.make_wave(start=0, duration=1, framerate=200)
ww.plot()
```

```python
spec = ww.make_spectrum()
spec.plot()
```

```python
spec.low_pass(cutoff=80)
spec.plot()
```
