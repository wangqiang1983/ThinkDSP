```python
from thinkdsp import PI2
class SawtoothSignal(thinkdsp.Sinusoid):
    def __init__(self,*parg, **kargs):
        super(SawtoothSignal,self).__init__(*parg, **kargs)
    def evaluate(self,ts):
        cycles = self.freq * ts +self.offset/PI2
        frac,_ = np.modf(cycles)
        ys = np.abs(frac - 0.5)
        
        return ys
```

```python
sig = SawtoothSignal(200)
```

```python
sig.plot()
```
