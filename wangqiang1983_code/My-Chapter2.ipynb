```python
wav = tri.make_wave(start=0,duration=0.5, framerate=10000)
wav.plot()
```

# 2 using the sig.make_wave(start, duration, framerate) to make a wave

```python
seg = wav.segment(start = 0, duration = tri.period*5)
seg.plot()
```

# 3 using wav.segment(start, duration) to make a segment

```python
spect = seg.make_spectrum()
spect.plot()
```
