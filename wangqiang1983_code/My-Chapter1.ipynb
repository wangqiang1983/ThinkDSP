```python
sin_sig = thinkdsp.SinSignal(freq=440, amp=2.0, offset = 0)
cos_sig = thinkdsp.CosSignal(freq=880, amp=1.5, offset = 0.003)
```

```python
sin_sig.plot()
thinkplot.config(xlabel='Time (s)',ylabel='Amp')
```

### sin(x)

```python
cos_sig.plot()
thinkplot.config(xlabel='Time (s)',ylabel='Amp')
```

### cos(x)

```python
mix = sin_sig + cos_sig
mix.plot()
thinkplot.config(xlabel='Time (s)',ylabel='Amp')
```

## sin(x) + cos(x)

```python
wave = mix.make_wave(duration=0.5, start=0, framerate=11025)  #duration=0.5, start=0, framerate=11025)
#wave.normalize()
#wave.apodize()
wave.plot()
```

# there are 5\*11025   sample points

```python
wave.make_audio()
```

```python
seg = wave.segment(start=0,duration=mix.period*6)
seg.plot()
```

```python
spec = wave.make_spectrum()
spec.plot()
thinkplot.config(xlabel='Frequency (Hz)')
```

```python
spec.low_pass(cutoff=800)
w = spec.make_wave()
seg = w.segment(start=0,duration=mix.period*3)
seg.plot()
```
