```python
sig1 = Sinc(200)
wav1 = sig1.make_wave(start=0,duration=0.02, framerate=10000)
wav1.plot()
```

```python
class MyUt(Sinusoid):
    def __init__(self,*parg, **kargs):
        super(MyUt,self).__init__(*parg, **kargs)
        
    def evaluate(self,ts):
        cycles = self.freq * ts +self.offset/PI2
        frac,_ = np.modf(cycles)
        frac+=0.0000000001
        ys = np.sign(frac)
        return ys
    
myut = MyUt()
myutwave = myut.make_wave()
myutwave.plot()
```

```python
convolved = np.convolve(wav1.ys,myutwave.ys)
convolved_wav = thinkdsp.Wave(convolved)
spec = convolved_wav.segment(duration=0.02)
spec.plot()
```
